```python
disease_df = pd.read_csv("obs_diabetes.csv") 
pop_df = pd.read_csv("pop_data_by_county.csv")
```

```python

#turned fips into float so we can merge on fips
disease_df["FIPS"] = disease_df["FIPS"].astype(float)
disease_df.head()
```

```python
states_df = pd.read_csv("states.csv")
states_df.head()
```

```python
disease_df = pd.merge(disease_df, states_df, how = "inner", left_on = "State", right_on = "abbreviation")
disease_df.head()
```

```python
disease_df = disease_df.drop("abbreviation", axis = 1)
disease_df.head()
```

```python
# pop_df['State'] = pop_df['State'].str.strip() #strip trailing white space from state name
county_data = pd.merge(disease_df, pop_df, how = "inner", left_on = "FIPS", right_on = "FIPS ")
county_data
```

```python
county_data.shape
```

```python
county_data = county_data.drop(["FIPS", "PCT_DIABETES_ADULTS08", "PCT_OBESE_ADULTS08",
                            "PCT_OBESE_ADULTS08", "PCT_HSPA15", "RECFAC09", "2010 Census Population",
                            "Population Estimate, 2011", "Population Estimate, 2012", "Population Estimate, 2014",
                           "Population Estimate, 2015", "Population Estimate, 2016"], axis = 1)
county_data.head()
```

```python
county_data = county_data.drop(["RECFAC14", "PCH_RECFAC_09_14", "RECFACPTH09", "RECFACPTH14", 
                               "PCH_RECFACPTH_09_14", "state_name",  "State_y",
                               "County_x",], axis = 1)
county_data.head()
```

```python
# county_data = county_data.drop("FIPS ", axis = 1)
```

```python
county_data.head()
```

```python
# county_data = pd.DataFrame({"State" : State_x, "County": County_x, "% Of Adults with Diabetes, 2013" : PCT_DIABETES_ADULTS13,
#                            "% Of Obese Adults, 2013" : PCT_OBESE_ADULTS13})

county_data = county_data.rename(columns={"State_x": "State" , "County_y": "County", "FIPS " : "FIPS",
                                         "PCT_DIABETES_ADULTS13": "% Of Adults with Diabetes, 2013",
                                            "PCT_OBESE_ADULTS13": "% Of Obese Adults, 2013" })
```

```python
county_data.head()
```

```python
county_data = county_data[['State','FIPS', 'County', '% Of Adults with Diabetes, 2013',
                          '% Of Obese Adults, 2013','Population Estimate, 2013' ]]


county_data['Population Estimate, 2013'] = pd.to_numeric(county_data['Population Estimate, 2013'].str.replace(',', ''))
county_data.head()
```

```python

d = {'State':'State', '% Of Adults with Diabetes, 2013':'% Of Adults with Diabetes, 2013',
     '% Of Obese Adults, 2013':'% Of Obese Adults, 2013', 'Population Estimate, 2013' : 'Population Estimate, 2013' }
gby_state = county_data.drop('FIPS', axis=1).groupby('State').agg({"% Of Adults with Diabetes, 2013" : 'mean', '% Of Obese Adults, 2013' : 'mean',
                                          'Population Estimate, 2013' : 'sum'   }).rename(columns=d)
gby_state.head()
# df=df.groupby('Name').agg({'Missed':'sum', 'Credit':'sum','Grade':'mean'}).rename(columns=d)
# print (df)
```
