```python
import pandas as pd
```

```python
wic_2013=pd.read_csv("wic_nslp_brk_2013.csv")
states=pd.read_csv("states.csv")
```

```python
wic_2013.head()
```

```python
# county_data = county_data[['State','FIPS', 'County', '% Of Adults with Diabetes, 2013',
#                           '% Of Obese Adults, 2013','Population Estimate, 2013' ]]


# county_data['Population Estimate, 2013'] = pd.to_numeric(county_data['Population Estimate, 2013'].str.replace(',', ''))
# county_data.head()

wic_2013[['State','WIC participants, FY 2013','National School Lunch Program participants, FY 2013',
                  'School Breakfast Program participants, FY 2013','Summer Food participants, FY 2013',
                   'State Population,  2013']]
wic_2013['WIC participants, FY 2013'] = pd.to_numeric(wic_2013['WIC participants, FY 2013'].str.replace(',', ''))

wic_2013['National School Lunch Program participants, FY 2013'] = pd.to_numeric(wic_2013['National School Lunch Program participants, FY 2013'].str.replace(',', ''))

wic_2013['School Breakfast Program participants, FY 2013'] = pd.to_numeric(wic_2013['School Breakfast Program participants, FY 2013'].str.replace(',', ''))

wic_2013['Summer Food participants, FY 2013'] = pd.to_numeric(wic_2013['Summer Food participants, FY 2013'].str.replace(',', ''))

wic_2013[ 'State Population,  2013'] = pd.to_numeric(wic_2013[ 'State Population,  2013'].str.replace(',', ''))
wic_2013.head()
```

```python
wic_2013.columns
```

```python
# disease_df = pd.merge(disease_df, states_df, how = "inner", left_on = "State", right_on = "abbreviation")
# disease_df.head()

wic_2013 = pd.merge(wic_2013, states, how = "inner", left_on = "State", right_on = "state_name")
wic_2013.head()
```

```python
# disease_df = disease_df.drop("abbreviation", axis = 1)
# # disease_df.head()

food_data = wic_2013.drop(["StateFIPS", "state_name", "State"], axis = 1)
food_data.head()
```

```python
food_data = food_data.rename(columns={"abbreviation" : "State", "WIC participants, FY 2013" : "WIC Participants", 
                                      "National School Lunch Program participants, FY 2013" : "Nat School Lunch Program Participants"
                                    , "School Breakfast Program participants, FY 2013" : "School Breakfast Program Participants"
                                     , "Summer Food participants, FY 2013" : "Summer Food Participants", 
                                      "State Population,  2013" : "State Population"})
food_data.head()
```

```python
food_data = food_data[["State", "WIC Participants", "Nat School Lunch Program Participants", 
                       "School Breakfast Program Participants", "Summer Food Participants", "State Population"]]
food_data.head()
```

```python
food_data=food_data.drop("State Population", axis=1)
food_data.head()
                
```

```python
st_di_pop=pd.read_csv("state_disease_pop_data.csv")
```

```python
st_di_pop.head()
```

```python
st_di_pop = st_di_pop.rename(columns={"% Of Adults with Diabetes, 2013":"% Adults with Diabetes",
                                      "% Of Obese Adults, 2013":"% Obese Adults","Population Estimate, 2013":"Population"})
st_di_pop.head()
```

```python
# st_di_pop=st_di_pop.drop("Population Estimate, 2013",axis=1)
# st_di_pop.head()
```

```python
all_data = pd.merge(food_data, st_di_pop, how = "inner", on = "State")
all_data.head()
```
