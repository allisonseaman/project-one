```python
# importing dependencies
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

```python
## oading WIC Participants
wic = pd.read_csv("all_2013_data.csv")
```

```python
# Divide population by 10 to make the charts readable
wic["Population / 10"] = wic["Population"] / 10
```

```python
# Find Top Ten WIC Participants by sorting by the largest number
top_ten_wic = wic.sort_values("WIC Participants", ascending = False)
```

```python
# Create a new data frame with top ten states
top_ten_wic_df = top_ten_wic[top_ten_wic.State.isin(['CA', 'TX', 'AK', 'OK', 'MS', 'AR', 'LA', 'KY', 'NM', 'GA'])]\
    [["State","WIC Participants", "Population", "Population / 10"]]
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(top_ten_wic_df))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,10))
p1 = plt.bar(x_axis, top_ten_wic_df["WIC Participants"], color = "black", alpha = 0.5, align = "center")
p2 = plt.bar(x_axis, top_ten_wic_df["Population / 10"], color = "green", alpha = 0.1, align = "center")
plt.xticks(tick_locations, top_ten_wic_df["State"], rotation = "horizontal")
plt.xlabel("State")
plt.ylabel("WIC Participants vs. Population")
plt.ylim(0, 4000000)
plt.legend((p1[0], p2[0]), ("WIC Participants", "Population"))
plt.show()
# plt.savefig("Wic_Participants_vs_Population_H.png")
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,5))
p1 = plt.bar(x_axis, top_ten_wic_df["WIC Participants"], color = "green", alpha = 0.5, align = "center")
plt.xticks(tick_locations, top_ten_wic_df["State"], rotation = "horizontal")
plt.xlabel("State")
plt.ylabel("WIC Participants")
plt.ylim(0, 1600000)
plt.legend(("WIC Participants", ))
plt.show()
# plt.savefig("Wic_Participants_H.png")
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,10))
plt.bar(x_axis, top_ten_wic_df["Population / 10"], color = "green", alpha = 0.5, align = "center")
plt.xticks(tick_locations, top_ten_wic_df["State"], rotation = "horizontal")
plt.xlabel("State")
plt.ylabel("Population / 10")
plt.ylim(0, 4000000)
plt.legend(("Population", ))
plt.show()
# plt.savefig("Population_H.png")
```

```python
# Loading Top Ten States for Diabetic and Obese Adults
data=pd.read_csv("all_2013_data.csv")
diabetes=pd.read_csv("top_ten_diabetes.csv")
obese=pd.read_csv("top_ten_obese.csv")
```

```python
# Create new columns with # of Obese and Diabetic Adults by Population
data["Number of Obese Adults"] = data["% Obese Adults"] / 100 * data["Population"]
data["Number of Diabetic Adults"] = data["% Adults with Diabetes"] / 100 * data["Population"]
```

```python
# Find Top Ten Lunch Program Participants by sorting by the largest number
obese = data.sort_values("Number of Obese Adults", ascending = False)
diabetes = data.sort_values("Number of Diabetic Adults", ascending = False)
```

```python
# Create a new data frame with top ten states
obese_ten = obese[obese.State.isin(["CA", "TX", "FL", "NY", "PA", "IL", "OH", "MI", "GA", "NC"])]\
    [["State", "Population", "Number of Obese Adults"]]
diabetes_ten = diabetes[diabetes.State.isin(["CA", "TX", "FL", "NY", "PA", "OH", "IL", "GA", "NC", "MI"])]\
    [["State", "Population","Number of Diabetic Adults"]]
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(diabetes_ten))
tick_locations = [value for value in x_axis]
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,10))
p1 = plt.bar(x_axis, diabetes_ten["Number of Diabetic Adults"], color='black', alpha=1, align="center")
p2 = plt.bar(x_axis, diabetes_ten["Population"], color='green', alpha=0.5, align="center")
plt.xticks(tick_locations, diabetes["State"], rotation="horizontal")
plt.xlabel("State")
plt.ylabel("Number of Diabetic Adults vs. Population")
plt.ylim(0, 40000000)
plt.legend((p1[0], p2[0]), ("Number of Diabetic Adults", "Population"))
plt.show()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(obese_ten))
tick_locations = [value for value in x_axis]
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,10))
p1 = plt.bar(x_axis, obese_ten["Number of Obese Adults"], color='black', alpha=1, align="center")
p2 = plt.bar(x_axis, obese_ten["Population"], color='green', alpha=0.5, align="center")
plt.xticks(tick_locations, obese["State"], rotation="horizontal")
plt.xlabel("State")
plt.ylabel("Number of Obese Adults vs. Population")
plt.ylim(0, 40000000)
plt.legend((p1[0], p2[0]), ("Number of Obese Adults", "Population"))
plt.show()
```
